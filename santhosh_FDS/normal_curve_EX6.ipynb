```python
import numpy as np
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

def generate_dataset(n):
    # Initialize empty lists for x and y values
    x = []
    y = []
    
    # Random coefficients for y = random_x1 * x1 + random_x2 * x2 + 1
    random_x1 = np.random.rand()
    random_x2 = np.random.rand()
    
    # Generate the dataset
    for i in range(n):
        x1 = i
        x2 = i / 2 + np.random.rand()  # Some random noise
        x.append([1, x1, x2])  # Append a row with 1, x1, and x2 (for x[0] = 1, x[1] = x1, x[2] = x2)
        y.append(random_x1 * x1 + random_x2 * x2 + 1)  # Linear function to generate y values
    
    return np.array(x), np.array(y)

# Generate the dataset with 200 data points
x, y = generate_dataset(200)

# Set plot parameters
mpl.rcParams['legend.fontsize'] = 12

# Create a 3D plot
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(projection='3d')

# Plot the 3D scatter plot with x1 and x2 as the axes and y as the value
ax.scatter(x[:, 1], x[:, 2], y, label='y', s=10)

# Set legend and view
ax.legend()
ax.view_init(45, 122)  # Set the view angle of the 3D plot

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a 3D axes object
ax = plt.axes(projection='3d')

# Create data for the line
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)

# Plot the 3D line
ax.plot3D(xline, yline, zline, 'gray')

# Create data for the scatter points
z_data = 15 * np.random.random(100)
x_data = np.sin(z_data) + 0.1 * np.random.randn(100)
y_data = np.cos(z_data) + 0.1 * np.random.randn(100)

# Plot the 3D scatter points
ax.scatter3D(x_data, y_data, z_data, c=z_data, cmap='Greens')

# Show the plot
plt.show()

```

```python
def f(x,y): 
    return np.sin(np.sqrt(x ** 2 + y ** 2)) 
x=np.linspace(-6, 6, 30) 
y = np.linspace(-6, 6,30) 
X,Y=np.meshgrid(x, y) 
Z=f(X,Y) 
fig= plt.figure() 
ax=plt.axes(projection='3d') 
ax.contour3D(X,Y,Z,50,cmap='binary') 
ax.set_xlabel('x') 
ax.set_ylabel('y') 
ax.set_zlabel('z');
```
