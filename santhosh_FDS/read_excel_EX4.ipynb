```python
import pandas as pd

# Create the DataFrame
data = {
    's.no': [1, 2, 3, 4, 5, 6, 7],
    'rollno': [100, 101, 102, 103, 104, 105, 106],
    'name': ['a', 'b', 'c', 'd', 'e', 'f', 'g'],
    'sub1': [91, 95, 92, 91, 89, 87, 89],
    'sub2': [89, 95, 94, 92, 81, 94, 93],
    'sub3': [87, 89, 87, 91, 92, 94, 92]
}

df = pd.DataFrame(data)

file_path = 'student_data.xlsx'
df.to_excel(file_path, index=False)

print(f"Excel file created at: {file_path}")
```
